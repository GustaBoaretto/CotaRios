```python
# Transformar o formato
df_tratado = df_tratado.melt(
    id_vars=["Data", "Maxima", "Minima", "Media"],  # Colunas fixas
    var_name="Dia",  # Nome para as colunas de Cota
    value_name="Cota"  # Nome para os valores
)

# Verificar o formato das datas e converter corretamente
df_tratado["Data"] = pd.to_datetime(df_tratado["Data"], format="%d/%m/%Y", dayfirst=False)

# Extrair o número do dia da coluna "Dia"
df_tratado["Dia_Num"] = df_tratado["Dia"].str.extract(r'(\d+)').astype(int)

# Adicionar o dia extraído à data base
df_tratado["Data_Completa"] = df_tratado["Data"] + pd.to_timedelta(df_tratado["Dia_Num"] - 1, unit="d")

# Exibir as primeiras linhas do DataFrame tratado
print(df_tratado.head())
```

```python
# Remover colunas desnecessárias
df_tratado = df_tratado.drop(columns=["Dia", "Dia_Num"])
```

```python
# Ordenar pela coluna Data
df_tratado = df_tratado.sort_values(by="Data_Completa")

# Excluir as colunas desnecessárias
df_tratado = df_tratado.drop(columns=["Data", "Maxima", "Minima", "Media"])

# Remover registros duplicados
df_tratado = df_tratado.drop_duplicates()

# Remover registros com dias inválidos (31 em meses que têm apenas 30 dias ou menos)
df_tratado = df_tratado[df_tratado["Data_Completa"].apply(
    lambda x: x.day <= pd.Timestamp(year=x.year, month=x.month, day=1).days_in_month
)]
```

```python
# Criar um índice de datas completo (sem saltos)
data_inicio = df_tratado["Data_Completa"].min()
data_fim = df_tratado["Data_Completa"].max()
datas_completas = pd.date_range(start=data_inicio, end=data_fim, freq='D')

# Criar um DataFrame com todas as datas e fazer o merge com o original
df_datas_completas = pd.DataFrame({"Data_Completa": datas_completas})
df_tratado = pd.merge(df_datas_completas, df_tratado, on="Data_Completa", how="left")


df_tratado
```

```python
import numpy as np

# Tratando Outliers
df_tratado = df_tratado.reset_index()
df_tratado['Data_Completa'] = pd.to_datetime(df_tratado['Data_Completa'])
df_tratado.set_index('Data_Completa', inplace=True)

#Cálculo do IQR (Interquartil)
Q1 = df_tratado['Cota'].quantile(0.25)
Q3 = df_tratado['Cota'].quantile(0.75)
IQR = Q3 - Q1

#Definindo limites para detecção de outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

#Substituindo outliers por NaN
df_tratado['Cota_sem_outlier'] = df_tratado['Cota'].where(
    (df_tratado['Cota'] >= limite_inferior) & (df_tratado['Cota'] <= limite_superior), np.nan)

#Interpolação linear para imputar valores
df_tratado['Cota_interpolada'] = df_tratado['Cota_sem_outlier'].interpolate(method='linear')

#Visualização opcional (antes e depois)
plt.figure(figsize=(14, 6))
plt.plot(df_tratado.index, df_tratado['Cota'], label='Original', alpha=0.5)
plt.plot(df_tratado.index, df_tratado['Cota_interpolada'], label='Interpolada (sem outliers)', color='green')
plt.legend()
plt.title('Interpolação com Remoção de Outliers')
plt.xlabel('Data')
plt.ylabel('Cota')
plt.grid(True)
plt.show()
```

```python
df_tratado['Cota'] = df_tratado['Cota_interpolada']
df_tratado.drop(columns=['Cota_sem_outlier', 'Cota_interpolada'], inplace=True)
df_tratado = df_tratado.reset_index()
```

```python
from statsmodels.tsa.stattools import kpss, adfuller


serie = df_tratado['Cota']

# Teste KPSS
stat_kpss, p_value_kpss, n_lags_kpss, crit_values_kpss = kpss(serie, regression='c', nlags="legacy")

print("### Teste KPSS ###")
print(f"KPSS Test Statistic: {stat_kpss}")
print(f"P-Valor: {p_value_kpss}")
print(f"Número de lags usados: {n_lags_kpss}")
print("Valores críticos:", crit_values_kpss)

if p_value_kpss > 0.05:
    print("A série é estacionária (KPSS).")
else:
    print("A série NÃO é estacionária (KPSS).")

print("\n//\n")

# Teste ADF
result_adf = adfuller(serie, autolag='AIC')

print("### Teste ADF ###")
print(f"ADF Test Statistic: {result_adf[0]}")
print(f"P-Valor: {result_adf[1]}")
print(f"Número de lags usados: {result_adf[2]}")
print(f"Número de observações usadas: {result_adf[3]}")
print("Valores críticos:")
for key, value in result_adf[4].items():
    print(f"  {key}: {value}")

if result_adf[1] <= 0.05:
    print("A série é estacionária (ADF).")
else:
    print("A série NÃO é estacionária (ADF).")
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import pandas as pd


# Copiando os dados
df_ETS = df_tratado.copy()

# Parâmetros
horizonte = 30

# Separando os últimos 30 dias como janela de previsão final (modelo nunca verá esses dados)
df_base = df_ETS[:-horizonte]
test_30dias = df_ETS[-horizonte:]  # conjunto de teste

# Definindo proporção de treino/teste
train_size = int(len(df_base) * 0.8)
train = df_base[:train_size]
test = df_base[train_size:]

# Treinamento 80/20 (sem os últimos 30 dias)
model_8020 = ExponentialSmoothing(train['Cota'],
                                  trend='add',
                                  damped_trend=True,
                                  seasonal='add',
                                  seasonal_periods=365,
                                  initialization_method='estimated',
                                  freq='D')
result_8020 = model_8020.fit(optimized=True)

# Previsão para o conjunto de teste (20%)
forecast_test = result_8020.forecast(steps=len(test))
forecast_test.index = test.index

# Avaliação no conjunto de teste
mse_test = mean_squared_error(test['Cota'], forecast_test)
rmse_test = np.sqrt(mse_test)
mape_test = mean_absolute_percentage_error(test['Cota'], forecast_test)

# Reajustando o modelo com todos os dados exceto os últimos 30 dias
model_final = ExponentialSmoothing(df_base['Cota'],
                                   trend='add',
                                   damped_trend=True,
                                   seasonal='add',
                                   seasonal_periods=365,
                                   initialization_method='estimated',
                                   freq='D')
result_final = model_final.fit(optimized=True)

# Prevendo os 30 últimos dias da série
forecast_30 = result_final.forecast(steps=horizonte)
forecast_30.index = test_30dias.index

# Avaliação nos 30 dias finais
mse_30 = mean_squared_error(test_30dias['Cota'], forecast_30)
rmse_30 = np.sqrt(mse_30)
mape_30 = mean_absolute_percentage_error(test_30dias['Cota'], forecast_30)

# Resultados
print("\n--- Avaliação no conjunto de teste (20%) ---")
print(f"MSE: {mse_test:.2f}")
print(f"RMSE: {rmse_test:.2f}")
print(f"MAPE: {mape_test:.2%}")

print("\n--- Avaliação na previsão dos últimos 30 dias ---")
print(f"MSE: {mse_30:.2f}")
print(f"RMSE: {rmse_30:.2f}")
print(f"MAPE: {mape_30:.2%}")

# Visualização
plt.figure(figsize=(14, 6))
plt.plot(train.index, train['Cota'], label='Treinamento', color='blue')
plt.plot(test.index, test['Cota'], label='Teste Real (20%)', color='green')
plt.plot(test.index, forecast_test, label='Previsto (20%)', linestyle='--', color='red')
plt.plot(test_30dias.index, test_30dias['Cota'], label='Real (Últimos 30 dias)', color='orange')
plt.plot(test_30dias.index, forecast_30, label='Previsto (Últimos 30 dias)', linestyle='--', color='darkorange')
plt.axvline(x=test.index[0], color='grey', linestyle=':', alpha=0.7, label='Início Teste')
plt.axvline(x=test_30dias.index[0], color='black', linestyle=':', alpha=0.7, label='Início dos 30 dias finais')
plt.title('Holt-Winters: Avaliação em Teste 20% + Últimos 30 Dias da Série')
plt.xlabel('Data')
plt.ylabel('Cota')
plt.legend()
plt.grid(True)
plt.show()

# Visualização apenas dos últimos 30 dias
plt.figure(figsize=(12, 5))
plt.plot(test_30dias.index, test_30dias['Cota'], label='Real (Últimos 30 dias)', color='red', marker='o')
plt.plot(test_30dias.index, forecast_30, label='Previsto (Últimos 30 dias)', linestyle='--', color='blue', marker='x')
plt.title('Comparativo Previsão vs Real - Últimos 30 dias')
plt.xlabel('Data')
plt.ylabel('Cota')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Copiando os dados e renomeando colunas
df_prophet = df_tratado.copy()
df_prophet = df_prophet.rename(columns={'Data_Completa': 'ds', 'Cota': 'y'})

horizonte = 30

# Separando os últimos 30 dias da base
df_base = df_prophet[:-horizonte]
test_30dias = df_prophet[-horizonte:]  # Conjunto de teste final

# Definindo 80/20 sem incluir os 30 últimos dias
train_size = int(len(df_base) * 0.8)
train = df_base[:train_size]
test = df_base[train_size:]

# Avaliação na base de teste (20%)
model_test = Prophet(yearly_seasonality=True)
model_test.fit(train)

# Criar datas futuras para previsão da base de teste
future_test = model_test.make_future_dataframe(periods=len(test), freq='D')
forecast_test = model_test.predict(future_test)

# Avaliação
mse_test = mean_squared_error(test['y'], forecast_test['yhat'][-len(test):])
rmse_test = np.sqrt(mse_test)
mape_test = mean_absolute_percentage_error(test['y'], forecast_test['yhat'][-len(test):])

# Previsão dos últimos 30 passos (fim da base)
model_final = Prophet(yearly_seasonality=True)
model_final.fit(df_base)

# Criando datas futuras para os 30 passos finais
future_30 = model_final.make_future_dataframe(periods=horizonte, freq='D')
forecast_30 = model_final.predict(future_30)

# Seleciona apenas as últimas 30 previsões
forecast_30_final = forecast_30.tail(horizonte)

# Avaliação dos últimos 30 dias
mse_30 = mean_squared_error(test_30dias['y'], forecast_30_final['yhat'])
rmse_30 = np.sqrt(mse_30)
mape_30 = mean_absolute_percentage_error(test_30dias['y'], forecast_30_final['yhat'])

# Resultados
print("\n--- Avaliação na base de teste (20%) ---")
print(f"MSE: {mse_test:.2f}")
print(f"RMSE: {rmse_test:.2f}")
print(f"MAPE: {mape_test:.2%}")

print("\n--- Avaliação nos últimos 30 dias ---")
print(f"MSE: {mse_30:.2f}")
print(f"RMSE: {rmse_30:.2f}")
print(f"MAPE: {mape_30:.2%}")

# Plotagem
plt.figure(figsize=(14, 6))
plt.plot(train['ds'], train['y'], label='Treinamento', color='blue')
plt.plot(test['ds'], test['y'], label='Teste Real (20%)', color='green')
plt.plot(forecast_test['ds'][-len(test):], forecast_test['yhat'][-len(test):], label='Previsto (Teste)', linestyle='--', color='red')
plt.plot(test_30dias['ds'], test_30dias['y'], label='Real (Últimos 30 dias)', color='orange')
plt.plot(forecast_30_final['ds'], forecast_30_final['yhat'], label='Previsto (Últimos 30 dias)', linestyle='--', color='darkorange')
plt.axvline(x=test['ds'].iloc[0], color='grey', linestyle=':', alpha=0.7, label='Início Teste')
plt.axvline(x=test_30dias['ds'].iloc[0], color='black', linestyle=':', alpha=0.7, label='Início 30 dias finais')
plt.title('Prophet: Avaliação em Teste 20% + Previsão dos Últimos 30 Dias')
plt.xlabel('Data')
plt.ylabel('Cota')
plt.legend()
plt.grid(True)
plt.show()

# Visualização apenas dos últimos 30 dias
plt.figure(figsize=(12, 5))
plt.plot(test_30dias['ds'], test_30dias['y'], label='Real (Últimos 30 dias)', color='red', marker='o')
plt.plot(forecast_30_final['ds'], forecast_30_final['yhat'], label='Previsto (Últimos 30 dias)', linestyle='--', color='darkorange', marker='x')
plt.title('Prophet - Comparativo Previsão vs Real - Últimos 30 dias')
plt.xlabel('Data')
plt.ylabel('Cota')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Criar lags para transformar em formato tabular
def create_lags(data, target="Cota", lag=0):
    df_lags = data.copy()

    # Criar colunas de lag
    for i in range(1, lag + 1 ):
        df_lags[f"{target}_lag_{i}"] = df_lags[target].shift(i)

    # Remover valores NaN gerados pelos lags
    df_lags.dropna(inplace=True)

    # Remover todas as colunas originais(tirando a data)
    df_lags = df_lags[[target] + [f"{target}_lag_{i}" for i in range(1, lag + 1)]]

    df_lags.rename(columns={target: "target"}, inplace=True)

    return df_lags
```

```python
def multi_step_forecast_df(final_model, last_known, start_date, steps=30, lag=6):
    forecasts = []
    current_lags = list(last_known[-lag:])

    for i in range(steps):
        # Criar input com os lags
        input_data = pd.DataFrame([current_lags], columns=[f"Cota_lag_{i+1}" for i in range(lag)])
        # Previsão
        pred = predict_model(final_model, data=input_data)
        next_val = pred["prediction_label"].iloc[0]
        forecasts.append(next_val)

        # Atualizar os lags
        current_lags = [next_val] + current_lags[:-1]

    # Gerar as datas futuras com base na data inicial
    date_range = pd.date_range(start=start_date, periods=steps, freq='D')

    # Retornar como DataFrame com as datas
    return pd.DataFrame({'Data': date_range, 'Previsao_Cota': forecasts}).set_index('Data')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input, Conv1D, MaxPooling1D
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt

# Função para preparar os dados para o modelo
def preparar_dados(df, coluna_cota, janela=30):
    scaler = MinMaxScaler(feature_range=(0, 1))
    df_scaled = scaler.fit_transform(df[coluna_cota].values.reshape(-1, 1))  # Normalização

    X, y = [], []
    for i in range(len(df_scaled) - janela):
        X.append(df_scaled[i:i + janela, 0])  # Sequência de entrada (janela)
        y.append(df_scaled[i + janela, 0])    # Próximo valor (alvo)

    X = np.array(X)
    y = np.array(y)

    # Reshape para entrada em 3D para LSTM
    X = X.reshape(X.shape[0], X.shape[1], 1)

    return X, y, scaler

# Preparar dados
janela = 30
X, y, scaler = preparar_dados(df_base, coluna_cota='Cota', janela=janela)

# Dividir os dados em treino, validação e teste
train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.1)
X_train, X_val, X_test = X[:train_size], X[train_size:train_size+val_size], X[train_size+val_size:]
y_train, y_val, y_test = y[:train_size], y[train_size:train_size+val_size], y[train_size+val_size:]

# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Modelo LSTM
model_lstm = Sequential()
model_lstm.add(Input(shape=(X_train.shape[1], 1)))
model_lstm.add(LSTM(units=50, return_sequences=False))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
model_lstm.fit(X_train, y_train, epochs=200, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stop])

# Previsão LSTM
y_pred_lstm = model_lstm.predict(X_test)
y_pred_lstm_rescaled = scaler.inverse_transform(y_pred_lstm)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Métricas LSTM
mse_lstm = mean_squared_error(y_test_rescaled, y_pred_lstm_rescaled)
rmse_lstm = np.sqrt(mse_lstm)
mape_lstm = mean_absolute_percentage_error(y_test_rescaled, y_pred_lstm_rescaled) * 100

# Modelo CNN + LSTM
model_cnn_lstm = Sequential()
model_cnn_lstm.add(Input(shape=(X_train.shape[1], 1)))
model_cnn_lstm.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model_cnn_lstm.add(MaxPooling1D(pool_size=2))
model_cnn_lstm.add(LSTM(units=50, return_sequences=False))
model_cnn_lstm.add(Dense(units=1))
model_cnn_lstm.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
model_cnn_lstm.fit(X_train, y_train, epochs=200, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stop])

# Previsão CNN + LSTM
y_pred_cnn_lstm = model_cnn_lstm.predict(X_test)
y_pred_cnn_lstm_rescaled = scaler.inverse_transform(y_pred_cnn_lstm)

# Métricas CNN + LSTM
mse_cnn_lstm = mean_squared_error(y_test_rescaled, y_pred_cnn_lstm_rescaled)
rmse_cnn_lstm = np.sqrt(mse_cnn_lstm)
mape_cnn_lstm = mean_absolute_percentage_error(y_test_rescaled, y_pred_cnn_lstm_rescaled) * 100

# Comparação
print(f"Erro Quadrático Médio (MSE) LSTM: {mse_lstm:.4f}")
print(f"Raiz do Erro Quadrático Médio (RMSE) LSTM: {rmse_lstm:.4f}")
print(f"Erro Percentual Médio Absoluto (MAPE) LSTM: {mape_lstm:.2f}%\n")

print(f"Erro Quadrático Médio (MSE) CNN + LSTM: {mse_cnn_lstm:.4f}")
print(f"Raiz do Erro Quadrático Médio (RMSE) CNN + LSTM: {rmse_cnn_lstm:.4f}")
print(f"Erro Percentual Médio Absoluto (MAPE) CNN + LSTM: {mape_cnn_lstm:.2f}%\n")

# Diferenças
print(f"Diferença de MSE: {abs(mse_lstm - mse_cnn_lstm):.4f}")
print(f"Diferença de RMSE: {abs(rmse_lstm - rmse_cnn_lstm):.4f}")
print(f"Diferença de MAPE: {abs(mape_lstm - mape_cnn_lstm):.2f}%")

# Preparando datas para plotagem
datas_plot = df_base['Data'].iloc[janela + train_size + val_size:].reset_index(drop=True)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(datas_plot, y_test_rescaled, label='Valor Real', color='blue')
plt.plot(datas_plot, y_pred_lstm_rescaled, label='Previsão LSTM', color='red', linestyle='dashed')
plt.plot(datas_plot, y_pred_cnn_lstm_rescaled, label='Previsão CNN + LSTM', color='green', linestyle='dashed')
plt.title('Previsão de Cotas com Modelos LSTM e CNN + LSTM')
plt.xlabel('Data')
plt.ylabel('Cota')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import numpy as np
import matplotlib.pyplot as plt

# Número de passos (dias)
passos = 30

# Função para previsão multi-step com janela deslizante
def previsao_multi_step(modelo, dados, janela=30, passos=30):
    previsoes = []
    entrada = dados[-janela:].reshape(1, janela, 1).copy()

    for i in range(passos):
        pred = modelo.predict(entrada, verbose=0)
        previsoes.append(pred[0][0])
        entrada = np.append(entrada[:, 1:, :], [[[pred[0][0]]]], axis=1)

    return scaler.inverse_transform(np.array(previsoes).reshape(-1, 1))

# Normaliza os dados de entrada
df_scaled = scaler.transform(df_base["Cota"].values.reshape(-1, 1))

# Gerar previsões multi-step com os dois modelos
previsoes_futuras_lstm = previsao_multi_step(model_lstm, df_scaled, janela=30, passos=passos)
previsoes_futuras_cnn_lstm = previsao_multi_step(model_cnn_lstm, df_scaled, janela=30, passos=passos)

# Seleciona os últimos 'passos' dias da base real para comparação
cotas_reais = df_futuro_real["Cota"].values[:passos]
datas_futuro = df_futuro_real["Data"].values[:passos]

# Avaliação dos modelos
mse_lstm_futuro = mean_squared_error(cotas_reais, previsoes_futuras_lstm)
rmse_lstm_futuro = np.sqrt(mse_lstm_futuro)
mape_lstm_futuro = mean_absolute_percentage_error(cotas_reais, previsoes_futuras_lstm) * 100

mse_cnn_lstm_futuro = mean_squared_error(cotas_reais, previsoes_futuras_cnn_lstm)
rmse_cnn_lstm_futuro = np.sqrt(mse_cnn_lstm_futuro)
mape_cnn_lstm_futuro = mean_absolute_percentage_error(cotas_reais, previsoes_futuras_cnn_lstm) * 100

# Exibição dos resultados
print(f"[LSTM] MSE: {mse_lstm_futuro:.4f} | RMSE: {rmse_lstm_futuro:.4f} | MAPE: {mape_lstm_futuro:.2f}%")
print(f"[CNN+LSTM] MSE: {mse_cnn_lstm_futuro:.4f} | RMSE: {rmse_cnn_lstm_futuro:.4f} | MAPE: {mape_cnn_lstm_futuro:.2f}%")

# Visualização
plt.figure(figsize=(12, 6))
plt.plot(datas_futuro, cotas_reais, label="Real", marker='o', color="blue")
plt.plot(datas_futuro, previsoes_futuras_lstm, label="LSTM", linestyle="--", marker='D', color="red")
plt.plot(datas_futuro, previsoes_futuras_cnn_lstm, label="CNN+LSTM", linestyle="--", marker='s', color="green")
plt.title(f"Previsão Multi-step para os Últimos {passos} Dias")
plt.xlabel("Data")
plt.ylabel("Cota")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
